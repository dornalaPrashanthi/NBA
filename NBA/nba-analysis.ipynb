```python
NBAdata=pd.read_csv('Seasons_Stats.csv')
```

```python
grp=NBAdata.groupby('Year')
x=grp['3P'].agg(np.mean)
plt.figure(figsize=(16,6))
plt.plot(x,'ro',color='r')
plt.xticks(rotation=90)
plt.title('Category wise Rating')
plt.xlabel('Year--->')
plt.ylabel('3P--->')
plt.show()
```
